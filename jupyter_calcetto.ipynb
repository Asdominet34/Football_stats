```python
import sqlite3
import pandas as pd
import os

# Percorso del file DB
db_path = r"C:\Users\Davide\OneDrive\Desktop\python\football_stats\calcetto_stats.db"
output_dir = r"C:\Users\Davide\OneDrive\Desktop\python\football_stats\json_output"

# Crea la cartella se non esiste
os.makedirs(output_dir, exist_ok=True)

# Connessione al database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Ottieni l'elenco delle tabelle
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

print("Tabelle trovate nel DB:")

# Per ogni tabella, salviamo in DataFrame e JSON
for table in tables:
    table_name = table[0]
    print(f"- {table_name}")

    # Leggi la tabella in un DataFrame
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)

    # Mostra anteprima del DataFrame
    print(df.head())

    # Salva in JSON
    json_path = os.path.join(output_dir, f"{table_name}.json")
    df.to_json(json_path, orient='records', indent=4)
    print(f"  → Salvata in: {json_path}\n")

# Chiudi la connessione
conn.close()

print("✅ Tutte le tabelle sono state esportate in JSON.")
```

```python
import sqlite3
import pandas as pd
import os

# Percorso del file .db
db_path = r"C:\Users\Davide\OneDrive\Desktop\python\football_stats\calcetto_stats.db"
excel_file = r"C:\Users\Davide\OneDrive\Desktop\python\football_stats\calcetto_completo.xlsx"

# Connessione al database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Recupera i nomi delle tabelle
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

print("Tabelle trovate nel DB:")

# Scrive tutte le tabelle in un file Excel con più fogli
with pd.ExcelWriter(excel_file, engine='xlsxwriter') as writer:
    for table in tables:
        table_name = table[0]
        print(f"- {table_name}")

        # Leggi la tabella in un DataFrame
        df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)

        # Scrivi su un foglio Excel (max 31 caratteri per nome foglio)
        df.to_excel(writer, sheet_name=table_name[:31], index=False)
        print(f"  → Foglio Excel aggiunto per: {table_name}")

# Chiudi la connessione
conn.close()
print(f"\n✅ Tutte le tabelle sono state salvate in: {excel_file}")
```
